```python

import yt_dlp

url = 'https://youtu.be/O9v10jQkm5c?si=kJHwQOx87tjb9EbZ'

ydl_opts = {
    'format': 'bestvideo+bestaudio/best', # Downloads best quality
    'outtmpl': '%(title)s.%(ext)s',        # Saves with video title
}

with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    ydl.download([url])
```

```python
import yt_dlp

url = 'https://youtu.be/iGVJ1oSHcEo?si=frtW9IQfSYcqYKBB'

ydl_opts = {
    'format': 'bestaudio/best',  # Download only the best audio quality
    'outtmpl': '%(title)s.%(ext)s', # Save with video title
    'postprocessors': [{        # Post-process to convert to MP3
        'key': 'FFmpegExtractAudio',
        'preferredcodec': 'mp3',
        'preferredquality': '192', # 192kbps is usually sufficient for RAG/transcription
    }],
}

with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    ydl.download([url])
```
